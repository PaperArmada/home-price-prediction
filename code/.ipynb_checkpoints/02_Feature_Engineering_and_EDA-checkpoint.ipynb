```python
features = df.drop(columns=['SalePrice', 'Id']).columns
features
```

```python
X = df[features]
y = df['SalePrice']
```

```python
X.isnull().sum()
```

```python
lr = LinearRegression()
```

```python
cross_val_score(lr, X, y).mean()
```

```python
lr.fit(X, y)
```

```python
df_test = pd.read_csv('../datasets/filtered_test_dataset.csv')
df_test.info()
```

```python
X_test = df_test[features].copy()
X_test.isnull().sum().sort_values(ascending=False)
```

```python
imputer = SimpleImputer()
imputer.fit(X)
```

```python
imputed_values = imputer.transform(X_test)
```

```python
imputed_X_test = pd.DataFrame(imputed_values, columns=X_test.columns)
```

```python
pred = lr.predict(imputed_X_test)
```

```python
lr.score(X, y)
```

```python
df_test['SalePrice'] = pred
df_test['SalePrice'].head()
```

```python
submission = df_test[['Id', 'SalePrice']]
submission.set_index('Id', inplace=True)
submission
```

```python
submission.to_csv('../outputs/mlr_4_submission.csv')
```
